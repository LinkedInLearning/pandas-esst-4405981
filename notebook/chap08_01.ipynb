```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
plt.style.use("ggplot")
```

```python
x = np.linspace(0,3*np.pi,100)
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.subplot(121)
plt.plot(x,y1)
plt.subplot(122)
plt.plot(x,y2)
```

```python
plt.subplot(221)
plt.plot(x,y1,"-r") #-
plt.subplot(222)
plt.plot(x,y2,"--b") #--
plt.subplot(223)
plt.plot(x,y1,"-.") #-.
plt.subplot(224)
plt.plot(x,y2,":g") #:
```
