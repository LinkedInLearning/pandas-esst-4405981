```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "名前":["相川","井伊","上野"],
    "年齢":[23,24,25],
    "所属":["営業課","開発課","総務課"]
})
```

```python
df2 = pd.DataFrame({
    "名前":["江藤","大野","柿沢"],
    "年齢":[36,37,39],
    "所属":["営業課","開発課","営業課"]
})
```

```python
df3 = pd.concat([df1,df2])
```

```python
df4 = pd.DataFrame({
    "名前":["江藤","大野","柿沢"],
    "役職":["チーフ","係長","課長"]
})
```

```python
df3
```

```python
df4
```

```python
df3.merge(df4)
```

```python
df3.merge(df4, how="inner")
```

```python
df3.merge(df4, how="outer")
```

```python
df3.merge(df4, how="left")
```

```python
df3.merge(df4, on="名前", how="outer")
```

```python
df4 = pd.DataFrame({
    "氏名":["江藤","大野","柿沢"],
    "役職":["チーフ","係長","課長"]
})
```

```python
df4
```

```python
df3.merge(df4, left_on="名前", right_on="氏名")
```

```python
df3.set_index("名前", inplace=True)
```

```python
df3
```

```python
df3.merge(df4, left_index=True, right_on="氏名")
```

```python
df4.set_index("氏名", inplace=True)
```

```python
df4
```

```python
df3.merge(df4, left_index=True, right_index=True)
```
