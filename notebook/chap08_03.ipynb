```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
plt.style.use("ggplot")
```

```python
left = np.array([1,2,3,4,5])
height = np.array([122,146,110,97,102])
labels = ["line1","line2","line3","line4","line5"]
```

```python
plt.title("Number of detections")
plt.xlabel("ID")
plt.ylabel("Number")
plt.bar(left, height, tick_label=labels)
```

```python
left = np.array([1,2,3,4,5])
height1 = np.array([122,146,110,97,102])
height2 = np.array([95,122,113,110,86])
height3 = np.array([123,176,133,100,115])
labels = ["line1","line2","line3","line4","line5"]
```

```python
plt.title("Number of detections")
plt.xlabel("ID")
plt.ylabel("Number")
plt.bar(left, height1, color="r", tick_label=labels)
plt.bar(left, height2, bottom=height1, color="b")
plt.bar(left, height3, bottom=height1 + height2, color="g")
```
