```python
import pandas as pd
```

```python
df = pd.read_csv("../data/members.csv")
```

```python
df["ID"]>300
```

```python
# IDが300より大きい会員だけを抽出
df[df["ID"]>300]
```

```python
# Member Typeが3の会員を抽出
df[df["Member Type"]==3]
```

```python
# Customer TypeがBusinessの会員を抽出
df[df["Customer Type"]=="Business"]
```

```python
# and は &
df[(df["Member Type"] == 3) & (df["Customer Type"] == "Business")]
```

```python
# or は |
df[(df["Member Type"] == 2) | (df["Member Type"] == 4)]
```

```python
# str.containsで抽出　メールアドレスに@blogを含む
df[df["Email"].str.contains("@blog")]
```

```python
# isin([])で抽出　メールアドレスに@blogを含む
df[df["Last Name"].isin(["Young"])]
```

```python
# isin([])で抽出　リストに複数指定
df[df["Last Name"].isin(["Young","Bell","Larson"])]
```

```python
# で始まる　str.startswith()
df[df["Phone"].str.startswith("(504")]
```

```python
# で終わる　str.endswith()
df[df["Email"].str.endswith(".au")]
```

```python
# ID列の最大値に一致するデータを抽出
df[df["ID"] == df["ID"].max()]
```
