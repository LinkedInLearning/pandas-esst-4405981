```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "名前":["相川","井伊","上野"],
    "年齢":[23,24,25],
    "所属":["営業課","開発課","総務課"]
})
```

```python
df2 = pd.DataFrame({
    "名前":["江藤","大野","柿沢"],
    "年齢":[36,37,39],
    "所属":["営業課","開発課","営業課"]
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], axis=0)
```

```python
pd.concat([df1,df2], ignore_index=True)
```

```python
df3 = pd.concat([df1,df2], ignore_index=True)
```

```python
df4 = df3.set_index("名前")
```

```python
df4
```

```python
df5 = pd.DataFrame({
    "名前":["江藤","大野","柿沢"],
    "役職":["チーフ","係長","課長"]
}).set_index("名前")
```

```python
df4
```

```python
df5
```

```python
pd.concat([df4,df5], axis=1)
```

```python
pd.concat([df4,df5], axis=1, join="outer")
```

```python
pd.concat([df4,df5], axis=1, join="inner")
```
