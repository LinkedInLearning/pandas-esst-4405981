```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
plt.style.use("ggplot")
```

```python
x = np.linspace(0,3*np.pi,50)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.ylim(-1.5, 1.5)
plt.title("sin(x)")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x,y,"c")
```
